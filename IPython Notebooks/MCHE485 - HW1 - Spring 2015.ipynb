## Problem 2.2

```python
# Let's look at a test case first
t = np.arange(0, 5.01, 0.01) # 0-5 seconds, with 0.01s spacing
y_d = 1.0                    # The step input amplitude

# Defint the spring and damping terms
k = (2 * np.pi)**2
c = 0.25

# First mass
m1 = 1.0
wn1 = np.sqrt(k / m1)
z1 = c / m1
wd1 = wn1 * np.sqrt(1 - z1**2)

decay_1 = np.exp(-z1 * wn1 * t)
decay_1_upper = y_d + np.exp(-z1*wn1*t) * y_d/(np.sqrt(1-z1**2))
decay_1_lower = y_d - np.exp(-z1*wn1*t) * y_d/(np.sqrt(1-z1**2))
step_resp_1 = y_d + np.exp(-z1*wn1*t)*(-y_d*np.cos(wd1*t) + (z1*y_d)/(np.sqrt(1-z1**2)) * np.sin(wd1*t))


# Second mass
m2 = 2.0
wn2 = np.sqrt(k / m1)
z2 = c / m2
wd2 = wn2 * np.sqrt(1 - z2**2)

decay_2 = np.exp(-z2 * wn2 * t)
decay_2_upper = y_d + np.exp(-z2*wn2*t) * y_d/np.sqrt(1-z2**2)
decay_2_lower = y_d - np.exp(-z2*wn2*t) * y_d/np.sqrt(1-z2**2)
step_resp_2 = y_d + np.exp(-z2*wn2*t)*(-y_d*np.cos(wd2*t) + (z2*y_d)/(np.sqrt(1-z2**2)) * np.sin(wd2*t))
```

```python
# Let's plot the two step responses

# Set the plot size - 3x2 aspect ratio is best
fig = plt.figure(figsize=(6, 4))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17, left=0.17, top=0.96, right=0.96)

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(), family='serif', fontsize=18)
plt.setp(ax.get_xmajorticklabels(), family='serif', fontsize=18)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':', color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
plt.xlabel('Time (s)', family='serif', fontsize=22, weight='bold', labelpad=5)
plt.ylabel('Position (m)', family='serif', fontsize=22, weight='bold', labelpad=10)

plt.plot(t, decay_1_upper, linewidth = 1.5, linestyle = '--', color = '#4daf4a')
plt.plot(t, decay_1_lower, linewidth = 1.5, linestyle = '--', color = '#4daf4a')
plt.plot(t, step_resp_1, linewidth=2, linestyle='-', label=r'$m = 1$ kg')
# plt.plot(t, step_resp_2, linewidth=2, linestyle='--', label=r'$m = 2$ kg')


# uncomment below and set limits if needed
# plt.xlim(0, 5)
plt.ylim(0, 2)

# Create the legend, then fix the fontsize
leg = plt.legend(loc='upper right', ncol = 1, fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext, family='serif', fontsize=20)

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# Uncommetn to save the figure as a high-res pdf in the current folder
# It's saved at the original 6x4 size
# plt.savefig('Problem2p2_StepResponses.pdf')

fig.set_size_inches(9, 6) # Resize the figure for better display in the notebook
```

## Problem 2.10
Let's use [SymPy](http://www.sympy.org/) to help us with the algebra of this problem.

```python
import sympy

# define sympy symbols for natural frequency and input frequency
wn, w = sympy.symbols('\omega_n \omega')

# Define the system transfer function
tf = wn**2 / (wn**2 - w**2)
```

```python
# We first sub 200 for w using g.subs(w,200), then solve g = -4/1.3 as g + 4/1.3 = 0
wn_sq = sympy.solve(tf.subs(w,200) + 4/1.3, wn**2)
```

```python
m = 5.0         # kg
g = 9.81        # gravity m/s^2

k = wn_sq[0] * 5.0 # sympy.solve returns a list, so we take the first element of it
```

```python
x_eq = m * g / k

print '\nThe static deflection of the mass is {:0.4f}m.\n'.format(float(x_eq))
```

## Problem 2.16
Let's again use SymPy to help us with the algebra.

```python
# define sympy symbols for natural frequency and input frequency
wn, w = sympy.symbols('\omega_n \omega')

# Define the system transfer function
tf = wn**2 / (wn**2 - w**2)

# We first sub 200 for w using g.subs(w,sympy.pi), then solve g = -1/2 as g + 1/2 = 0
wn_sq = sympy.solve(tf.subs(w, sympy.pi) + 1.0/2, wn**2)

natural_freq = sympy.sqrt(wn_sq[0]) # sympy.solve returns a list, so we take the first element of it
```

```python
print '\nThe natural frequency is {:0.4f} rad/s.\n'.format(float(natural_freq))
```

```python
# We could also solve for wn directly.

# We first sub 200 for w using g.subs(w,sympy.pi), then solve g = -1/2 as g + 1/2 = 0
natrual_freq = sympy.solve(tf.subs(w, sympy.pi) + 1.0/2, wn)

print '\nThe natural frequency is {:0.4f} rad/s.\n'.format(float(natural_freq))
```
