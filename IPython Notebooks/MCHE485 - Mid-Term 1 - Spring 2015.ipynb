# Problem 1

```python
m = 1.0                        # kg
k1 = 2.0 * np.pi               # N/m
k2 = 2.0 * np.pi               # N/m
mu = 0.1                       # Friction coefficient
g = 9.81                       # gravity

wn = np.sqrt((k1+k2) / m)      # natural frequency (rad/s)

z1 = 0.0                       # damping ratio
z2 = 0.2                       # damping ratio 2
z3 = 0.7                       # damping ratio 3
```

```python
# Define the system as a series of 1st order ODES (beginnings of state-space form)
def eq_of_motion_viscous(w, t, p):
    """
    Defines the differential equations for the viscously damped spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m, k1, k2, c, mu, g]
    """
    x, x_dot = w
    m, k1, k2, c, mu, g = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x_dot,
             (-(k1+k2) * x - c * x_dot) / m]
    return sysODE
```

```python
# Set up simulation parameters 

# ODE solver parameters
abserr = 1.0e-9
relerr = 1.0e-9
max_step = 0.01
stoptime = 5.0
numpoints = 5001

# Create the time samples for the output of the ODE solver
t = np.linspace(0.0, stoptime, numpoints)

# Initial conditions
x_init = 0.0                        # initial position
x_dot_init = 5.0                    # initial velocity

# Pack the initial conditions into arrays 
x0 = [x_init, x_dot_init]
```

```python
# Undamped (z=0) Simluation
c = 2.0 * z1 * wn * m
# Pack the parameters and initial conditions into arrays 
p = [m, k1, k2, c, mu, g]

# Call the ODE solver.
resp_viscous_un = odeint(eq_of_motion_viscous, x0, t, args=(p,), atol=abserr, rtol=relerr,  hmax=max_step)


# z = 0.2 Simulation
c = 2.0 * z2 * wn * m
wd = wn * np.sqrt(1 - z2**2)

# Pack the parameters and initial conditions into arrays 
p = [m, k1, k2, c, mu, g]

# Call the ODE solver.
resp_viscous_0p2 = odeint(eq_of_motion_viscous, x0, t, args=(p,), atol=abserr, rtol=relerr,  hmax=max_step)

# Also define the exponential decay envelop
amp = np.sqrt(x0[0]**2 + ((z2*wn*x0[0] + x0[1])/wd)**2)
decay_env_0p2 = amp * np.exp(-z2*wn*t)


# z = 0.7 Simulation
c = 2.0 * z3 * wn * m
wd = wn * np.sqrt(1 - z3**2)

# Pack the parameters and initial conditions into arrays 
p = [m, k1, k2, c, mu, g]

# Call the ODE solver.
resp_viscous_0p7 = odeint(eq_of_motion_viscous, x0, t, args=(p,), atol=abserr, rtol=relerr,  hmax=max_step)
```

```python
# Define the system as a series of 1st order ODES (beginnings of state-space form)
def eq_of_motion_fric(w, t, p):
    """
    Defines the differential equations for the coupled spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m, k1, k2, c, mu, g]
    """
    x, x_dot = w
    m, k1, k2, c, mu, g = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x_dot,
             (-(k1+k2) * x - mu * m * g * np.sign(x_dot)) / m]
    return sysODE
```

# Problem 2

Let's use the closed-form, steady-state solution we developed in lecture:

Assume:

$ \quad f(t) = \bar{f} e^{i \omega t} $

Then, the solution $theta(t)$ should have the form:

$ \quad x(t) = \left| \theta \right| e^{i \omega t} $



Substituting this assumed solution into the equations of motion and solving for $ \bar{\theta}$:

$ \quad \left| \bar{\theta} \right| = \frac{\bar{f}}{m l} \left( \frac{1}{\left(\omega_n^2 - \omega^2\right) + \left(2 i \zeta \omega \omega_n \right)} \right) $

where

$ \quad \omega_n = \sqrt{\frac{g}{l}} \quad \mbox{and} \quad 2\zeta\omega_n = \frac{c}{m} $

This can also be written in magnitude and phase form:

$ \quad \left| \bar{\theta} \right| = \frac{\bar{f}}{m l} \left( \frac{1}{\sqrt{\left(\omega_n^2 - \omega^2\right)^2 + \left(2 \zeta \omega \omega_n \right)^2}} \right) $

and

$ \quad \phi = \tan^{-1}\left({\frac{2 \zeta \omega_n \omega}{\omega_n^2 - \omega^2}}\right) $



```python
m = 1.0
g = 9.81
l = 2.0

wn = np.sqrt(g / l)

# Set up input parameters
wun = np.linspace(0,4*wn,500)          # Frequency range for freq response plot, 0-5wn with 500 points in-between
w = np.linspace(0,4*wn,500)            # Frequency range for freq response plot, 0-5wn with 500 points in-between

# Let's examine a few different damping ratios
z = 0.0
mag_normal_un = 1/(m*l*np.sqrt((wn**2 - w**2)**2 + (2*z*w*wn)**2))
phase_un = -np.arctan2((2*z*w*wn),(wn**2 - w**2)) * 180/np.pi

# Let's mask the phase discontinuity, so it isn't plotted
pos = np.where(np.abs(m*l*mag_normal_un) >= 25)
phase_un[pos] = np.nan
wun[pos] = np.nan



z = 0.2
mag_normal_0p2 = 1/(m*l*np.sqrt((wn**2 - w**2)**2 + (2*z*w*wn)**2))
phase_0p2 = -np.arctan2((2*z*w*wn),(wn**2 - w**2)) * 180/np.pi


z = 0.7
mag_normal_0p7 = 1/(m*l*np.sqrt((wn**2 - w**2)**2 + (2*z*w*wn)**2))
phase_0p7 = -np.arctan2((2*z*w*wn),(wn**2 - w**2)) * 180/np.pi
```

```python
# Set up the normalized frequency range
w = np.linspace(0,20,2000)            # Normalize Freq, 0-20 with 2000 points in-between

# Define the normalized transfer function for several different damping ratios
z = 0.0;
mag_normal_un = np.sqrt((1+(2*z*w)**2)/((1-w**2)**2+(2*z*w)**2))
phase_un = (np.arctan2(2*z*w,1) - np.arctan2((2*z*w),(1-w**2)))*180/np.pi

# Let's mask the discontinuity in the undamped phase response
# Let's mask the discontinuity, so it isn't plotted
pos = np.where(np.abs(1-w) <= 1e-2)
phase_un[pos] = np.nan

z = 0.2;
mag_normal_0p2 = np.sqrt((1+(2*z*w)**2)/((1-w**2)**2+(2*z*w)**2))
phase_0p2 = (np.arctan2(2*z*w,1) - np.arctan2((2*z*w),(1-w**2)))*180/np.pi

z = 0.7;
mag_normal_0p7 = np.sqrt((1+(2*z*w)**2)/((1-w**2)**2+(2*z*w)**2))
phase_0p7 = (np.arctan2(2*z*w,1) - np.arctan2((2*z*w),(1-w**2)))*180/np.pi
```
