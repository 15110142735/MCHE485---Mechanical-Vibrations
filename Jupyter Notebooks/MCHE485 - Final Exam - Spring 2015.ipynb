```python
# Define the system pararmeters
k1 = 2*(2*np.pi)**2           # Spring constant (N/m)
m1 = 1.75                     # Sprung/main mass (kg)
m2 = 0.25                     # rotating mass (kg)
m3 = 0.5                      # vibration absorber mass (kg)
k2 = k1 / 4.0                 # vibration absorber spring constant (N/m)
b = m2 / (m1 + m2)            # mass ratio    
wn = np.sqrt(k1 / (m1 + m2))  # natural frequency (rad/s)
l = 0.1                       # Eccentricity

z = 0.05                      # Damping Ratio
c = 2*z*wn*(m1 + m2)          # Select c based on desired amping ratio
```

Let's first look at the undamped eigenvalues and eigenvectors.

```python
M = np.array([[m1 + m2, 0],
              [0, m3]])

C = np.array([[c, 0],
              [0, 0]])

K = np.array([[k1 + k2, -k2],
              [-k2, k2]])

eigenvals, eigenvects = linalg.eigh(K,M)

print '\n'
print 'The resulting eigenalues are {:.2f} and {:.2f}.'.format(eigenvals[0], eigenvals[1])
print '\n'
print 'So, the two natrual frequencies are {:.2f}rad/s and {:.2f}rad/s.'.format(np.sqrt(eigenvals[0]), np.sqrt(eigenvals[1]))
print '\n'
print 'The first eigenvector is ' + str(eigenvects[:,0]) + '.'
print '\n'
print 'The second eigenvector is ' + str(eigenvects[:,1]) + '.'
print '\n'
```

```python
# Define the sytsem matrices - We'll use state-space form
A = np.array([[0, 1.0, 0, 0],
              [-(k1+k2) / (m1 + m2), -c / (m1 + m2), k2 / (m1 + m2), 0],
              [0, 0, 0, 1.0],
              [k2 / m3, 0, -k2 / m3, 0]])
```

```python
eigenvals_ss, eigenvects_ss = linalg.eig(A)

print '\n'
print 'The resulting eigenvalues are {:.4}, {:.4}, {:.4}, and {:.4}.'.format(eigenvals_ss[0], eigenvals_ss[1], eigenvals_ss[2], eigenvals_ss[3])
print '\n'

print 'So, the resulting natural frequencies are {:.4}rad/s and {:.4}rad/s.'.format(np.abs(eigenvals_ss[2]), np.abs(eigenvals_ss[0]))
print '\n'
```

```python
# make 1st entry real
eigvect1_ss = eigenvects_ss[:,0] * np.exp(-1.0j * np.angle(eigenvects_ss[0,0]))
eigvect2_ss = eigenvects_ss[:,2] * np.exp(-1.0j * np.angle(eigenvects_ss[0,2]))

# scale to match the undamped
eigvect1_ss *= (eigenvects[0,1] / eigvect1_ss[0])
eigvect2_ss *= (eigenvects[0,0] / eigvect2_ss[0])

print '\n'
print 'The first eigevector is '
print np.array_str(eigvect1_ss, precision=4, suppress_small=True)
print '\n'
print 'The second eigevector is '
print np.array_str(eigvect2_ss, precision=4, suppress_small=True)
print '\n'
```

```python
w = np.linspace(0,14,12000)

# We'll need to keep all our arrays imaginary
X = np.zeros((len(w),4)) + 0j

I = np.eye(4) + 0j
A_bar = A + 0j

# This is (I - w A)^-1 * F
for ii, freq in enumerate(w):
    F = np.array([0.0 + 0j, (l*b*freq**2)/(m1 + m2), 0.0, 0.0])
    X[ii,:] = np.dot(linalg.inv(freq * 1j * I - A_bar), F)

# Now, get the magnitude of the response
X_magnitude = np.abs(X)
```

```python
# Define the system pararmeters
m = 1.0                     # Sprung/main mass (kg)
k = 0.5 * (2 * np.pi)**2    # Spring constant (N/m)
mu = 0.23                   # Friction coefficient
g = 9.81                    # gravity
wn = np.sqrt(2 * k / m)     # natural frequency (rad/s)


z = 0.1                     # Damping Ratio
c = 2*z*wn*(m)              # Select c based on desired amping ratio  

wd = wn * np.sqrt(1 - z**2) # Damped natural frequency (rad/s)
```

```python
# Define the system as a series of 1st order ODES (beginnings of state-space form)
def eq_of_motion_viscous(w, t, p):
    """
    Defines the differential equations for the viscously damped spring-mass system.

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m, k, c, mu, g]
    """
    x, x_dot = w
    m, k, c, mu, g = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x_dot,
             (-2 * k * x - c * x_dot) / m]
    return sysODE


# Define the system as a series of 1st order ODES (beginnings of state-space form)
def eq_of_motion_fric(w, t, p):
    """
    Defines the differential equations for the spring-mass system with friction

    Arguments:
        w :  vector of the state variables:
                  w = [x, x_dot]
        t :  time
        p :  vector of the parameters:
                  p = [m, k, c, mu, g]
    """
    x, x_dot = w
    m, k, c, mu, g = p

    # Create sysODE = (x', x_dot'):
    sysODE = [x_dot,
             (-2* k * x - mu * m * g * np.sign(x_dot)) / m]
    return sysODE
```

```python
# Set up simulation parameters 

# ODE solver parameters
abserr = 1.0e-9
relerr = 1.0e-9
max_step = 0.01
stoptime = 5.0
numpoints = 5001

# Create the time samples for the output of the ODE solver
t = np.linspace(0.0, stoptime, numpoints)

# Initial conditions
x_init = 1.0                        # initial position
x_dot_init = -5.0                    # initial velocity

# Pack the initial conditions into arrays 
x0 = [x_init, x_dot_init]

# Pack the parameters
p = [m, k, c, mu, g]
```

Let's first simulate and plot the response for the viscously-damped system.

```python
# Call the ODE solver.
resp_viscous = odeint(eq_of_motion_viscous, x0, t, args=(p,), atol=abserr, rtol=relerr,  hmax=max_step)

# Also define the exponential decay envelop
amp = np.sqrt(x0[0]**2 + ((z*wn*x0[0] + x0[1])/wd)**2)
decay_env = amp * np.exp(-z*wn*t)
```

```python
# Define the system pararmeters
m1 = 1.0                     # kg
m2 = 1.1                     # kg
k = (2*np.pi)**2             # Spring constant (N/m)
c = 0.1                      # Damping coefficient
```

Now, let's look at the forced response.

Using $M$ and $K$, we want to solve:

$ \quad \left[K - \omega^2 M\right]\bar{X} = \bar{F} $ 

for $\bar{X}$. To do so, we need to take the inverse of $\left[K - \omega^2 M\right]$.

$ \quad \bar{X} = \left[K - \omega^2 M\right]^{-1}\bar{F} $ 

```python
T = 1.0

F = [T, 0.0]

w = np.linspace(0.01,12,1200)
X = np.zeros((len(w),2))

# This is (K-w^2 M)^-1 * F
for ii, freq in enumerate(w):
    X[ii,:] = np.dot(linalg.inv(K - freq**2 * M), F)

# Let's mask the discontinuity, so it isn't plotted
pos = np.where(np.abs(X[:,0]) >= 5)
X[pos,:] = np.nan
w[pos] = np.nan
```
